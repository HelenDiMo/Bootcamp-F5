```python
# DETECTANDO VARIABLES

edad = int(input("Ingrese su edad: "))
frecuencia_cardiaca = int(input("Ingrese su frecuencia cardiaca: "))
temperatura = float(input("Ingrese su temperatura: "))
sintomas_infecciosos = input("¿Tiene síntomas infecciosos? (Sí/No): ")
vacunas = input("¿Tiene vacunas? (Sí/No): ")
emergencia_critica = input("¿Está en emergencia crítica? (Sí/No): ")

# CODIGOS DE COLOR

ROJO = "\033[91m"
VERDE = "\033[92m"
AMARILLO = "\033[93m"
NARANJA_REAL = "\033[38;5;208m"
RESET = "\033[0m"

# CLASIFICACIÓN DE PRIORIDAD
print()
print("===========================")
print("CLASIFICACION DE PRIORIDAD")
print("===========================")
print()

if emergencia_critica == "Sí" or temperatura > 41:
  prioridad = "Roja"
  print(f"{ROJO}¡PRIORIDAD ROJA!{RESET}")

elif temperatura >= 39 and frecuencia_cardiaca > 100:
  prioridad = "Naranja"
  print(f"{NARANJA_REAL}¡PRIORIDAD NARANJA!{RESET}")

elif edad < 12 or edad > 65:
  prioridad = "Amarilla"
  print(f"{AMARILLO}¡PRIORIDAD AMARILLA!{RESET}")

else:
  prioridad = "Verde"
  print(f"{VERDE}¡PRIORIDAD VERDE!{RESET}")

print()

# PROTOCOLO DE BIOSEGURIDAD

if sintomas_infecciosos and not vacunas:
  destino = "BOX DE AISLAMIENTO"
  print("Diríjase al BOX DE AISLAMIENTO")
else:
  destino = "SALA DE ESPERA GENERAL"
  print("Diríjase a la SALA DE ESPERA GENERAL")

# RESUMEN FINAL PACIENTE

print()
print("===========================")
print("Resumen final del paciente:")
print("===========================")
print()
print("Edad:", edad)
print("Frecuencia cardiaca:", frecuencia_cardiaca)
print("Temperatura", temperatura)
print("Síntomas infecciosos:", sintomas_infecciosos)
print("Vacunas:", vacunas)
print("Emergencia crítica:", emergencia_critica)
print(f"Destino: {destino}")
print(f"Resultado de la evaluación: {prioridad}")
```
