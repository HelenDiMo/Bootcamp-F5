```python
import numpy as np

#Listas

a = [1, 2, 3, 4, 5]
b = [6, 7, 8, 9, 10]

array_a = np.array(a)
array_b = np.array(b)

# Sumatorio
print(array_a + array_b)


# Características
print("Tipo de objeto", type(a))
print("Tipo de objeto", type(array_a))
print("Tipo de datos", array_a.dtype)
print("Forma del array_a", array_a.shape)
print("Número tortal de elementos en el array_a_: ", len(array_a))
```

```python
# Creacion de un array en 2D

ar2 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])

print(ar2)
```

```python
#Array de solo ceros

ar0 = np.zeros(5)

print(ar0)
```

```python
# Array solo de 1

ar1 = np.ones((4, 2))

print(ar1)
```

```python
# Con rango (como range() de Python, pero devuelve un array)

print(ar2)
print()
np.arange(1, 10, 2) # Start, Stop, Cada cuanto queremos que avance
```

```python
# Random - Crea un array de 2 filas y 3 columnas y selecciona números random del 0 al 1

np.random.random((2, 3))
```

```python
# Full

arfull = np.full(4, 10)
print(arfull)
```
