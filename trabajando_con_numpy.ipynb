CREACION DE CAFETERIA DIGITAL

```python
# CREACION DE LA MATRIZ
# [cafés vendidos, croissants vendidos]

ventas = np.array([
    [10,5],#lunes
    [15,8],#martes
    [12,6],#miercoles
    [20,12],#jueves
    [25,15],#viernes
])

print("Nuestra tabla de ventas: ")
print(ventas)
```

```python
ventas[0]
croissants_vendidos = ventas[:,1]
print(croissants_vendidos)
```

```python
ventas_inicio_semana = ventas[0:2, :]
print(ventas_inicio_semana)
```

```python
ventas_ajuste = ventas + 0.5
print("Precios ajustados: ")
print(ventas_ajuste)
```

```python
# BROADCASTING

a = np.array([1, 2, 3])
b = 2

result = a + b
print(result)
# suma 2 a cada elemento de la fila
```

```python
arr = np.array([1, 2, 3])
matriz = np.array([[1, 2, 3], [4, 5, 6]])

result = arr + matriz
print(result)
```

```python
#2 cafes + 1 = 5€
#1 cafe + 1 croissant = 3€

#Código Sistemas Lineales
# A array de lo que compramos
A = np.array([
    [2, 1],#Lunes
    [1, 1] #Martes
])
#De lo que se paga
B = np.array([5, 3])

solucion = np.linalg.solve(A, B)
print(solucion)
#
```
