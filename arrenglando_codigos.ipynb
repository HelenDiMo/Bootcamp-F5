```python
def calcular_coste_distancia(km):
  precio = km * 0.5
  return precio # ERROR 1: No devuelve nada (none)

def aplicar_descuento(total):
  if total > 100:
    descuento = 20
    return descuento # ERROR 2: Devuelve el descuento, no el precio restado
  return total

def monstrar_ticket(final):
  # ERROR 3: La función no recibe lo que cree que recibe
  print(f"El coste final de su envío es: {final}")

# Programa Principal

distancia = int(input("Ingrese la distancia del envío: "))
fase1 = calcular_coste_distancia(distancia)
fase2 = aplicar_descuento(fase1)
monstrar_ticket(fase2)
```

```python
def alta_usuario():
    nombre = input("Introduce tu nombre: (Escriba 'salir' para salir del menú): ")
    if nombre.lower() == "salir":
        return None
    nombre_limpio = nombre.strip().title()
    return nombre_limpio

print("--- SISTEMA DE REGISTRO V1 ---")

usuarios = []

while True:
    nombre_limpio = alta_usuario()
    if nombre_limpio is None:
        break

    email = input("Introduce tu email: ")

    if "@" not in email or "." not in email:
        print("Email inválido")
    else:
        usuarios.append((nombre_limpio, email))
        print (f"Registrando a {nombre_limpio} con correo {email}...")
        print ("Usuario guardado con éxito.")

    opcion = input(" ¿Desea registrar a otro? (s/n): ")

    if opcion.lower() == "n":
        break

print("Fin del programa.")
print(f"Has registrado a {len(usuarios)}")
```
